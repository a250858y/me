```python
import pandas as pd

df = pd.read_csv("world_happiness_report.csv")

# 相関を調べたい列をまとめて抽出
cols = [
    "Happiness Score",
    "Economy (GDP per Capita)",
    "Family",
    "Health (Life Expectancy)",
    "Freedom",
    "Trust (Government Corruption)"
]

subset = df[cols]

# 相関行列を計算
correlation_matrix = subset.corr()

# 結果を表示
print(correlation_matrix)
```
